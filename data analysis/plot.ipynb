```python
def get_index_from_json(filepath):

    with open(filepath,"r") as f:
        data = json.loads(f.readline())
        ind_key = data["filenames"][0]
        ind = data[ind_key]
        # print(len(ind))
        
    return ind
```

```python
from matplotlib_venn import venn2
from matplotlib.cbook import flatten
from matplotlib import pyplot as plt
import numpy as np

figure, (ax1,ax2) = plt.subplots(1, 2, figsize=(12,5))

t_int=(153420,62376,108438)
f_int=(157350,1813401,3632775)


data = [t_int,f_int]
max_area = max(map(sum, data))

def set_venn_scale(vd, ax, true_area, reference_area=max_area):
    sx = np.sqrt(float(reference_area)/true_area)
    sy = max(vd.radii)*2
    ax.set_xlim(-sx, sx)
    ax.set_ylim(-sy, sy)


vd = venn2(data[0], set_colors=('#5858FA', '#D74E3B'), set_labels = ('Plausible\n streamlines\n from rCOMMIT', 'Plausible\n streamlines\n from rSIFT'), ax=ax1, subset_label_formatter=lambda x: f"{(x/10000000):1.1%}" )
lbl = vd.get_label_by_id("10")
x, y = lbl.get_position()
lbl.set_position((x-0.3, y))  

rbl = vd.get_label_by_id("01")
x, y = rbl.get_position()
rbl.set_position((x+0.3, y)) 

set_venn_scale(vd, ax1, sum(data[0]))


vd = venn2(data[1], set_colors=('#3E64AF', '#3EAF5D'),set_labels = ('Implausible\n streamlines\n from rCOMMIT', 'Implausible\n streamlines\n from rSIFT'), ax=ax2, subset_label_formatter=lambda x: f"{(x/10000000):1.1%}")
rbl = vd.get_label_by_id("Implausible\n streamlines\n from rCOMMIT")
x, y = rbl.get_position()
rbl.set_position((x-0.3, y-0.3)) 

set_venn_scale(vd, ax2, sum(data[1]))

# axes[1,2].axis('off')
figure.tight_layout(pad=0.1)
plt.title('Venn Diagram of Plausible Streamlines and Implausible Streamlines between rCOMMIT and rSIFT')
plt.show()
```
